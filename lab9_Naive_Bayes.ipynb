# Conditional Probabilities and Independence

- **Conditional Probability** of event A given event B is defined as:
  $
  \Pr[A \mid B] = \frac{\Pr[A \cap B]}{\Pr[B]}
  $
  when $\Pr[B] \neq 0$.

-  Therefore
  $
  \Pr[A \cap B] = \Pr[A] \Pr[B \mid A]  = \Pr[B] \Pr[A \mid B]
  $

- Events A and B are called **independent** if:
  $
  \Pr[A \cap B] = \Pr[A] \Pr[B]
  $

- If events A and B are independent, then:
  $
  \Pr[A \mid B] = \Pr[A]
  $

# Basic Properties

- $\Pr[A \cup B] = \Pr[A] + \Pr[B] - \Pr[A \cap B]$  
  *(Addition Rule)*

- $\Pr\left[\bigcup_{i=1}^{n} A_i\right] \leq \sum_{i=1}^{n} \Pr[A_i]$  
  *(Union Bound)*

- $\Pr[A \mid B] = \frac{\Pr[B \mid A] \Pr[A]}{\Pr[B]}$  
  *(Bayes' Formula)*

- $\Pr\left[\bigcap_{i=1}^{n} A_i\right] = \Pr[A_1] \Pr[A_2 \mid A_1] \Pr[A_3 \mid A_1 \cap A_2] \ldots \Pr[A_n \mid \bigcap_{i=1}^{n-1} A_i]$  
  *(Chain Rule)*

- $\Omega = A_1 \cup A_2 \cup \ldots \cup A_n$, where $A_i \cap A_j = \emptyset$ for $i \neq j \implies$  
  $\Pr[B] = \sum_{i=1}^{n} \Pr[B \mid A_i] \Pr[A_i]$  
  *(Total Probability Theorem)*

# Random Variables

- A **random variable** (random function) is a function $X : \Omega \to \mathbb{R}$ such that for every real interval $I \subset \mathbb{R}$, the preimage is an event, i.e., $X^{-1}(I) \in \mathcal{F}$ (always true for discrete probability spaces).

- **Consequence:** If $(\Omega, \mathcal{F}, \Pr)$ is a discrete probability space, $X : \Omega \to \mathbb{R}$ is a random variable, and $f : \mathbb{R} \to \mathbb{R}$ is a function, then $f(X) : \Omega \to \mathbb{R}$ is a random variable.

- The **distribution function** of a discrete random variable is the function:  
  $x \mapsto \Pr[X = x] = \Pr[X^{-1}(x)]$.

- The **joint distribution function** of discrete random variables $X, Y$ is the function:  
  $(x, y) \mapsto \Pr[X = x, Y = y] = \Pr[X^{-1}(x) \cap Y^{-1}(y)]$.

- Random variables $X, Y : \Omega \to \mathbb{R}$ are **independent** if:  
  $\Pr[X = x, Y = y] = \Pr[X = x] \Pr[Y = y]$ for every $x, y \in \mathbb{R}$.

- A sequence of random variables is called **independent and identically distributed** if they are mutually independent and have the same distribution function.

- A **multidimensional random variable** is a function $X : \Omega \to \mathbb{R}^n$, for which its projections $X_i : \Omega \to \mathbb{R}$, $X_i(\omega) := \text{Proj}_i X(\omega)$, $i = 1, 2, \ldots, n$ are random variables.

- The **distribution function** of a discrete multidimensional random variable is the function:  
  $(x_1, x_2, \ldots, x_n) \mapsto \Pr[X = (x_1, x_2, \ldots, x_n)] = \Pr[X^{-1}((x_1, x_2, \ldots, x_n))] = \Pr[X_1^{-1}(x_1) \cap X_2^{-1}(x_2) \cap \ldots \cap X_n^{-1}(x_n)] = \Pr[X_1 = x_1, X_2 = x_2, \ldots, X_n = x_n]$.

- **Property:** Let $X : \Omega \to \mathbb{R}^n$ be a random variable. Then $(X(\Omega), X(\mathcal{F}), \Pr)$ is a probability space.

# Principle of Maximum Likelihood

- Given a sequence of $m$ independent and identically distributed random variables $X_1, X_2, \ldots, X_m$ with a distribution function $\Pr[X = x \mid \theta]$, which depends on the parameter $\theta$.

- We have observed (measured) the corresponding values $x_1, x_2, \ldots, x_m$ for the sequence of random variables $X_1, X_2, \ldots, X_m$.

- The likelihood of having made the corresponding observation is:  
  $L(\theta) = \Pr[X_1 = x_1, X_2 = x_2, \ldots, X_m = x_m \mid \theta] = \prod_{i=1}^{m} \Pr[X_i = x_i \mid \theta]$

- We obtain the maximum likelihood by finding the value of the parameter $\theta$ for which the likelihood $L(\theta)$ is maximal. That is,  
  $\hat{\theta} = \arg \max_{\theta} L(\theta)$

# Maximizing Likelihood for Binomial Distribution

- Assume a binomial distribution function $B(n, p)$ for $m$ independent and identically distributed random variables $X_1, X_2, \ldots, X_m$ with observations $x_1, x_2, \ldots, x_m$. That is, $\Pr[X = x \mid p] = \binom{n}{x} p^x (1 - p)^{n-x}$.

- Let $f_k$ be the frequency of value $k$ among $x_1, x_2, \ldots, x_m$. We seek:
  $$\hat{p} = \arg \max_p L(p) = \arg \max_p \log L(p) = \arg \max_p \sum_{k=1}^{n} f_k \left( \log \binom{n}{k} + k \log p + (n-k) \log (1-p) \right)$$

- The derivative:
  $$\frac{\partial \log L(p)}{\partial p} = \sum_{k=1}^{n} \left( \frac{k f_k}{p} - \frac{(n-k) f_k}{1-p} \right) = 0$$

- Solving gives:
  $$(1-p) \sum_{k=1}^{n} k f_k = p \sum_{k=1}^{n} (n-k) f_k$$

- Therefore:
  $$\hat{p} = \frac{1}{nm} \sum_{i=1}^{m} x_i$$

To derive $\hat{p} = \frac{1}{nm} \sum_{i=1}^{m} x_i$ from the equation 

$$(1-p) \sum_{k=1}^{n} k f_k = p \sum_{k=1}^{n} (n-k) f_k,$$

follow these steps:

1. **Expand the Right Side:**

   The right side can be expanded as:
   $$p \sum_{k=1}^{n} (n-k) f_k = p \left( n \sum_{k=1}^{n} f_k - \sum_{k=1}^{n} k f_k \right).$$

2. **Simplify the Equation:**

   Substitute the expanded form into the equation:
   $$(1-p) \sum_{k=1}^{n} k f_k = p \left( n \sum_{k=1}^{n} f_k - \sum_{k=1}^{n} k f_k \right).$$

3. **Combine Terms:**

   Rearrange the terms to isolate $\sum_{k=1}^{n} k f_k$:
   $$\sum_{k=1}^{n} k f_k = \frac{p n \sum_{k=1}^{n} f_k}{1-p+p} = p n \sum_{k=1}^{n} f_k.$$

4. **Relate to Observations:**

   Recognize that $\sum_{k=1}^{n} k f_k$ is the total number of successes across all trials, which is equivalent to $\sum_{i=1}^{m} x_i$.

5. **Solve for $\hat{p}$:**

   Substitute back to find $\hat{p}$:
   $$\hat{p} = \frac{\sum_{i=1}^{m} x_i}{n \sum_{k=1}^{n} f_k}.$$

6. **Normalize by Total Trials:**

   Since $\sum_{k=1}^{n} f_k = m$, the total number of trials, we have:
   $$\hat{p} = \frac{1}{nm} \sum_{i=1}^{m} x_i.$$

This shows how the maximum likelihood estimate $\hat{p}$ is derived as the average number of successes per trial.

$$(1-p) \sum_{k=1}^{n} k f_k = p \left( n \sum_{k=1}^{n} f_k - \sum_{k=1}^{n} k f_k \right).$$

1. **Distribute $p$ on the Right Side:**

   $$p \left( n \sum_{k=1}^{n} f_k - \sum_{k=1}^{n} k f_k \right) = p n \sum_{k=1}^{n} f_k - p \sum_{k=1}^{n} k f_k.$$

2. **Combine Like Terms:**

   Move $p \sum_{k=1}^{n} k f_k$ to the left side:

   $$(1-p) \sum_{k=1}^{n} k f_k + p \sum_{k=1}^{n} k f_k = p n \sum_{k=1}^{n} f_k.$$

3. **Factor Out $\sum_{k=1}^{n} k f_k$:**

   $$\sum_{k=1}^{n} k f_k = p n \sum_{k=1}^{n} f_k.$$

4. **Solve for $\sum_{k=1}^{n} k f_k$:**

   Since $\sum_{k=1}^{n} f_k = m$, the total number of trials, we have:

   $$\sum_{k=1}^{n} k f_k = \sum_{i=1}^{m} x_i,$$

   where $\sum_{i=1}^{m} x_i$ is the total number of successes.

5. **Substitute Back to Find $\hat{p}$:**

   $$\hat{p} = \frac{\sum_{i=1}^{m} x_i}{n m}.$$

# Bernoulli Document Model

- Let a vocabulary $V = \{t_1, t_2, \ldots, t_M\}$ be given.

- For each document, we associate an $M$-dimensional vector of zeros and ones $d = (e_1, e_2, \ldots, e_M)$, where $e_i = 1$ if the term $t_i$ appears in the document and $e_i = 0$ otherwise. That is, $\mathbb{X} = \{0,1\}^M$.

- We assume $U_i : \mathbb{X} \rightarrow \{0,1\}$ are mutually independent random variables with Bernoulli distribution, such that $U_i$ gives us the $i$-th projection of the elements of $\mathbb{X}$.

- In this case:

  $$
  \text{Pr}[d] = \text{Pr}[(e_1, e_2, \ldots, e_M)] = \text{Pr}[U_1 = e_1, U_2 = e_2, \ldots, U_M = e_M] = \prod_{i=1}^{M} \text{Pr}[U_i = e_i]
  $$

- We seek the most probable class $c$ given that we have a document $d$. That is, we seek $c_{MAP} = \arg \max_{c \in C} \text{Pr}[c \mid d]$

- MAP = maximum a posteriori

- $\text{Pr}[c \mid d] = \frac{\text{Pr}[d \mid c] \, \text{Pr}[c]}{\text{Pr}[d]}$

- $\text{Pr}[d \mid c] = \text{Pr}[(e_1, e_2, \ldots, e_M) \mid c] = \prod_{i=1}^{M} \text{Pr}[U_i = e_i \mid c]$

- $c_{MAP} = \arg \max_{c \in C} \left( \text{Pr}[c] \prod_{i=1}^{M} \text{Pr}[U_i = e_i \mid c] \right)$

- $c_{MAP} = \arg \max_{c \in C} \left( \log \text{Pr}[c] + \sum_{i=1}^{M} \log \text{Pr}[U_i = e_i \mid c] \right)$

# Estimating Parameters Using the Maximum Likelihood Principle

- $N$ - number of documents in $\mathbb{D}$

- $N_c$ - number of documents in $\mathbb{D}$ of class $c$

- $N_{c,t}$ - number of documents in $\mathbb{D}$ of class $c$ in which the term $t$ appears

- $\text{Pr}[c] \approx \frac{N_c}{N}$

- $\text{Pr}[U_i = 1 \mid c] \approx \frac{N_{c,t_i}}{N_c} \approx \frac{N_{c,t_i} + 1}{N_c + 2}$

- $\text{Pr}[U_i = 0 \mid c] = 1 - \text{Pr}[U_i = 1 \mid c]$

# Algorithms for Naive Bayes Classifier Using Bernoulli Document Model

```plaintext
TrainBernoulliNB(C, D)
1   V <- EXTRACT_VOCABULARY(D)
2   N <- COUNT_DOCS(D)
3   for each c in C do
4       Nc <- COUNT_DOCS_IN_CLASS(D, c)
5       prior[c] <- Nc/N
6       for each t in V do
7           Nct <- COUNT_DOCS_IN_CLASS_CONTAINING_TERM(D, c, t)
8           condprob[t][c] <- (Nct + 1)/(Nc + 2)
9   return V, prior, condprob

ApplyBernoulliNB(C, V, prior, condprob, d)
1   Vd <- EXTRACT_TERMS_FROM_DOC(V, d)
2   for each c in C do
3       score[c] <- log prior[c]
4       for each t in V do
5           if t in Vd then
6               score[c] += log(condprob[t][c])
7           else
8               score[c] += log(1-condprob[t][c])
9   return argmax(c in C, score[c])
```

# Example of Distribution of a Discrete Random Variable

- Sample space: all possible outcomes when tossing $n$ coins.

- Random variable $X$: number of heads.

- **Binomial distribution**: $B(n, p)$

  $$
  \text{Pr}[X = k] = \binom{n}{k} p^k (1 - p)^{n-k}
  $$

- **Generalization**: Multinomial distribution $M(n, l, p_1, p_2, \ldots, p_l)$

  - Toss $n$ dice with $l$ sides.

  - Random variables: $X_1, X_2, \ldots, X_l$ - number of times side â€œ$i$â€ appears.

  $$
  \text{Pr}[X_1 = k_1, X_2 = k_2, \ldots, X_l = k_l] = \frac{n!}{k_1!k_2!\ldots k_l!} p_1^{k_1} p_2^{k_2} \ldots p_l^{k_l}
  $$

  when

  $$
  \sum_{i=1}^{l} k_i = n, \quad \sum_{i=1}^{l} p_i = 1.
  $$

# Maximizing Likelihood in Multinomial Distribution

- Assume a binomial function for the distribution $M(n, l, p_1, p_2, \ldots, p_l)$ of $m$ i.i.d. joint random variables $X^{(1)}, X^{(2)}, \ldots, X^{(m)}$, where $X^{(i)} = (X_1^{(i)}, X_2^{(i)}, \ldots, X_l^{(i)})$ with observations $\mathbf{x}^{(1)}, \mathbf{x}^{(2)}, \ldots, \mathbf{x}^{(m)}$, where $\mathbf{x}^{(i)} = (x_1^{(i)}, x_2^{(i)}, \ldots, x_l^{(i)})$. That is,

  $$
  \text{Pr}[X_1^{(i)} = x_1^{(i)}, X_2^{(i)} = x_2^{(i)}, \ldots, X_l^{(i)} = x_l^{(i)}] = \frac{n!}{x_1^{(i)}!x_2^{(i)}!\ldots x_l^{(i)}!} p_1^{x_1^{(i)}} p_2^{x_2^{(i)}} \ldots p_l^{x_l^{(i)}}
  $$

- We seek:

  $$
  \hat{\mathbf{p}} = \arg \max_{\mathbf{p}} L(\mathbf{p}) = \arg \max_{\mathbf{p}} \log L(\mathbf{p}) =
  $$

  $$
  = \arg \max_{\mathbf{p}} \sum_{i=1}^{m} \log \text{Pr}[X_1^{(i)} = x_1^{(i)}, X_2^{(i)} = x_2^{(i)}, \ldots, X_l^{(i)} = x_l^{(i)}] =
  $$

  $$
  = \arg \max_{\mathbf{p}} \sum_{i=1}^{m} \left( \log \frac{n!}{x_1^{(i)}!x_2^{(i)}!\ldots x_l^{(i)}!} + x_1^{(i)} \log p_1 + x_2^{(i)} \log p_2 + \ldots + x_l^{(i)} \log p_l \right)
  $$

- **Problem**: If $p_i \rightarrow \infty$, then $L(\mathbf{p}) \rightarrow \infty$.

- **Solution**: We need to find $\hat{\mathbf{p}} = \arg \max_{\mathbf{p}} \log L(\mathbf{p})$ with the constraint $\sum_{i=1}^{l} p_i = 1$.

- $$\hat{p}_i = \frac{\sum_{j=1}^{m} x_i^{(j)}}{nm}$$

# Multinomial Document Model

- Let a vocabulary $V = \{t_1, t_2, ..., t_M\}$ be given.

- For each document, we assign an $M$-dimensional vector of natural numbers:  
  $d = (f_1, f_2, ..., f_M)$, where $f_i$ represents the count of occurrences of term $t_i$ in the document.  
  That is, $\mathbb{X} = \mathbb{N}^M$.

- This representation of documents is called **Bag of Words**.

- Assuming that the document's length is $n = f_1 + ... + f_M$,  
  we have a multinomial distribution. Specifically:

  $$
  \text{Pr}[d] = \text{Pr}[(f_1,f_2, ..., f_M)] = K_d \prod_{i=1}^M \text{Pr}[t_i]^{f_i}, 
  $$
  where $K_d = \frac{n!}{f_1!f_2!...f_M!}$.

- For an arbitrary (variable) document length, the coefficient $K_d$ is multiplied by  
  the probability $\text{Pr}[|d| = n]$ â€” the probability of the given document $d$ having a length $n$.

- We are looking for the most likely class $c$ given that we have document $d$. That is, we are searching for:

  $$
  c_{MAP} = \operatorname{argmax}_{c \in C} \text{Pr}[c \mid d]
  $$

- By Bayes' theorem:

  $$
  \text{Pr}[c \mid d] = \frac{\text{Pr}[d \mid c] \cdot \text{Pr}[c]}{\text{Pr}[d]}
  $$

- The probability $\text{Pr}[d \mid c]$ can be expanded as:

  $$
  \text{Pr}[d \mid c] = \text{Pr}[(f_1, f_2, \ldots, f_M) \mid c] = K_d \prod_{i=1}^M \text{Pr}[t_i \mid c]^{f_i}
  $$

- Therefore, the maximum a posteriori class is:

  $$
  c_{MAP} = \operatorname{argmax}_{c \in C} \text{Pr}[c] \prod_{i=1}^M \text{Pr}[t_i \mid c]^{f_i}
  $$

- Taking the logarithm to simplify computation:

  $$
  c_{MAP} = \operatorname{argmax}_{c \in C} \log \text{Pr}[c] + \sum_{i=1}^M f_i \log \text{Pr}[t_i \mid c]
  $$

- Further simplified to:

  $$
  c_{MAP} = \operatorname{argmax}_{c \in C} \log \text{Pr}[c] + \sum_{k=1}^n \log \text{Pr}[t_{d_k} \mid c]
  $$

# Estimation of Parameters Using the Principle of Maximum Likelihood

- $N$ - the total number of documents in $\mathcal{D}$.

- $N_c$ - the number of documents in $\mathcal{D}$ that belong to class $c$.

- $T_{c,t}$ - the total occurrences of term $t$ in the documents in $\mathcal{D}$ belonging to class $c$.

- $$\text{Pr}[c] \approx \frac{N_c}{N}$$

- $$\text{Pr}[t_i \mid c] \approx \frac{T_{c, t_i}}{\sum_{t' \in V} T_{c, t'}} \approx \frac{T_{c, t_i} + 1}{\sum_{t' \in V} T_{c, t'} + |V|}$$

# Algorithms for Naive Bayes Classifier Using Multinomial Document Model

### TrainMultinomialNB(C, D)
1. $V \leftarrow \text{EXTRACT\_VOCABULARY}(D)$  
2. $N \leftarrow \text{COUNT\_DOCS}(D)$  
3. **for each** $c$ **in** $C$ **do**  
4. &nbsp;&nbsp;&nbsp;&nbsp;$N_c \leftarrow \text{COUNT\_DOCS\_IN\_CLASS}(D, c)$  
5. &nbsp;&nbsp;&nbsp;&nbsp;$\text{prior}[c] \leftarrow N_c / N$  
6. &nbsp;&nbsp;&nbsp;&nbsp;$\text{text}_c \leftarrow \text{CONCATENATE\_TEXT\_OF\_ALL\_DOCS\_IN\_CLASS}(D, c)$  
7. &nbsp;&nbsp;&nbsp;&nbsp;**for each** $t$ **in** $V$ **do**  
8. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$Tc[t] \leftarrow \text{COUNT\_TOKENS\_OF\_TERM}(\text{text}_c, t)$  
9. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**for each** $t$ **in** $V$ **do**  
10. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\text{condprob}[t][c] \leftarrow (\text{Tc}[t] + 1) / \big(\sum (\text{Tc}[t'] + 1) \ \text{for} \ t' \in V \big)$  
11. **return** $V$, $\text{prior}$, $\text{condprob}$  

---

### ApplyMultinomialNB(C, V, prior, condprob, d)
1. $W \leftarrow \text{EXTRACTTOKENSFROMDOC}(V, d)$  
2. **for each** $c$ **in** $C$ **do**  
3. &nbsp;&nbsp;&nbsp;&nbsp;$\text{score}[c] \leftarrow \log(\text{prior}[c])$  
4. &nbsp;&nbsp;&nbsp;&nbsp;**for each** $t$ **in** $W$ **do**  
5. &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\text{score}[c] \mathrel{+}= \log(\text{condprob}[t][c])$  
6. **return** $\text{argmax}(c \ \text{in} \ C, \ \text{score}[c])$

````python
train_spam = ['send us your password', 'review our website', 'send your password', 'send us your account']
train_ham = ['Your activity report','benefits physical activity', 'the importance vows']
test_spam = ['renew your password', 'renew your vows']
test_ham = ['benefits of our account', 'the importance of physical activity']
````

````python
import pandas as pd
````

````python
data = pd.DataFrame({
    'text': train_spam + test_spam + train_ham + test_ham,
    'label': [1] * (len(train_spam) + len(test_spam)) + [0] * (len(train_ham) + len(test_ham))
})
data
````

````python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
#remove the punctuations and stopwords
import string

def text_process(text):

    text = text.translate(str.maketrans('', '', string.punctuation))
    text = [word for word in text.split() if word.lower() not in stopwords.words('english')]

    return " ".join(text)

text = data['text'].apply(text_process)
````

````python
text
````

````python
from collections import Counter

total_counts = Counter()
for i in range(len(text)):
    for word in text.values[i].split(' '):
        total_counts[word] += 1

print("Total words in data set: ", len(total_counts))
````

````python
vocab = sorted(total_counts, key=total_counts.get, reverse=True)
print(vocab)
````

````python
vocab_size = len(vocab)
word2idx = {}
#print vocab_size
for i, word in enumerate(vocab):
    word2idx[word] = i
    
word2idx
````

````python
import numpy as np
def text_to_vector(text):
    word_vector = np.zeros(vocab_size)
    for word in text.split(" "):
        if word2idx.get(word) is None:
            continue
        else:
            word_vector[word2idx.get(word)] += 1
    return np.array(word_vector)
````

````python
word_vectors = np.zeros((len(text), len(vocab)), dtype=np.int_)
for i, t in enumerate(text):
    word_vectors[i] = text_to_vector(t)

print(word_vectors, word_vectors.shape)
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(word_vectors, data['label'], test_size=0.5, random_state=111)
````

````python
from sklearn.naive_bayes import MultinomialNB
````

````python
mnb = MultinomialNB(alpha=0.2)
````

````python
pred_scores_word_vectors = []

mnb.fit(X_train, y_train)
print(mnb.predict(X_test))
````

A sentiment analysis job about the problems of each major U.S. airline. Twitter data was scraped from February of 2015 and contributors were asked to first classify positive, negative, and neutral tweets, followed by categorizing negative reasons (such as "late flight" or "rude service").

#### The Goal: Create a Machine Learning Algorithm that can predict if a tweet is positive, neutral, or negative. In the future we could use such an algorithm to automatically read and flag tweets for an airline for a customer service agent to reach out to contact.

## Features and Label

````python
data = df[["airline_sentiment", "text"]]
````

````python
data.head()
````

````python
y = df["airline_sentiment"]
X = df["text"]
````

### Train Test Split

````python
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=101
)
````

## Vectorization

````python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
````

````python
countVec = CountVectorizer(stop_words="english")
````

````python
countVec.fit(X_train)
````

````python
X_train_countVec = countVec.transform(X_train)
X_test_countVec = countVec.transform(X_test)
````

````python
X_train_countVec
````

````python
tfidf = TfidfVectorizer(stop_words="english")
````

````python
tfidf.fit(X_train)
````

````python
X_train_tfidf = tfidf.transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
````

````python
X_train_tfidf
````

**DO NOT USE .todense() for such a large sparse matrix!!!**

## Naive Bayes

````python
from sklearn.naive_bayes import MultinomialNB
````

````python
nb_cv = MultinomialNB()
nb_cv.fit(X_train_countVec, y_train)
````

````python
nb_tfidf = MultinomialNB()
nb_tfidf.fit(X_train_tfidf, y_train)
````

## Performance Evaluation

````python
from sklearn.metrics import (
    ConfusionMatrixDisplay,
    classification_report,
    confusion_matrix,
)
````

````python
def report(model):
    preds = model.predict(X_test_tfidf)
    print(classification_report(y_test, preds))
    cm = confusion_matrix(y_test, preds, labels=model.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot()
````

````python
print("Count vectorization model")
report(nb_cv)
````

````python
print("Tf-idf model")
report(nb_tfidf)
````

### Finalizing a PipeLine for Deployment on  New Tweets

If we were satisfied with a model's performance, we should set up a pipeline that can take in a tweet directly.

````python
from sklearn.pipeline import Pipeline
````

````python
pipe = Pipeline([("tfidf", TfidfVectorizer()), ("naiveBase", MultinomialNB())])
````

````python
pipe.fit(df["text"], df["airline_sentiment"])
````

````python
new_tweet = ["I really liked the flight "]
print(pipe.predict(new_tweet))
````
